# Dynamical Pion EFT Hamiltonian

The full Hamiltonian for the Dynamical Pion EFT is
$$𝐻_{𝐷𝜋} =𝐻_{free} +𝐻_𝐶 +𝐻_{𝐶_{𝐼^2}} +𝐻_𝜋 +𝐻_{𝑁𝜋} $$
Where the free nucleon Hamiltonian is 
$$H_{\text{free}} = -h \sum_{\langle x,y \rangle} \sum_{\sigma} \left( a^\dagger_\sigma(x)a_\sigma(y) + \text{h.c.} \right) + 6h \sum_{x} \sum_{\sigma} N_\sigma(x)$$
The contact terms are on-site interactions:
$$\tilde{H}_C = \frac{C}{8} \sum_{i} \sum_{\sigma, \sigma'} (1 - Z_{\sigma, i})(1 - Z_{\sigma', i})$$
and 
$$H_{CI^2} = \frac{C_{I^2}}{2} \sum_{x} : \left[ \sum_{\sigma} N_\sigma^2 - 6N_{\uparrow p}N_{\uparrow n} + 2N_{\uparrow p}N_{\downarrow p} \dots - 4(a^\dagger_{\uparrow p} a_{\downarrow p} a^\dagger_{\downarrow n} a_{\uparrow n} + \text{h.c.}) \right] :$$

The free pion Hamiltonian $H_\pi$ is:
$$H_\pi = \frac{a_L^3}{2} \sum_{x} \sum_{I} [\Pi_I^2(x) + (\nabla \pi_I (x))^2 + m_{\pi}^2 \pi_I^2 (x)]$$

Finally, the pion-nucleon interaction is accounted for by $H_{N\pi}$, which is split into the axial-vector term, $H_{\text{AV}}$,and the Weinberg-Tomozawa term, $H_{\text{WT}}$:
$$H_{N\pi} = H_{\text{AV}} + H_{\text{WT}}$$
with 
$$H_{\text{AV}} = \frac{g_A}{2f_\pi} \sum_{x} \sum_{I, S} a^\dagger(x) \tau_I \sigma_S \partial_S \pi_I(x) a(x)$$
and 
$$H_{\text{WT}} = \frac{1}{4f_\pi^2} \sum_{x} \sum_{I_1, I_2, I_3} \epsilon_{I_1 I_2 I_3} \pi_{I_2}(x) \Pi_{I_3}(x) a^\dagger(x) \tau_{I_1} a(x)$$


# Encoding the Hamiltonian

For the free fermion Hamiltonian:
Qubit Mapping of Hopping Terms 
Depending on the direction of the hop ($x, y, z$), the Jordan-Wigner string differs. Here $\tilde{a}$ represents the qubit-mapped fermion operator.X-axis: $\tilde{h}_{x,\sigma}(i,j) \to \tilde{a}^\dagger_\sigma(i)\tilde{a}_\sigma(j) + \text{h.c.}$ Y-axis: $\tilde{h}_{y,\sigma}(i,j) \to (\tilde{a}^\dagger_\sigma(i)\tilde{a}_\sigma(j) + \text{h.c.}) \cdot i \tilde{\mu}(i)\bar{\mu}(j)$ Z-axis: $\tilde{h}_{z,\sigma}(i,j) \to (\tilde{a}^\dagger_\sigma(i)\tilde{a}_\sigma(j) + \text{h.c.}) \cdot i \tilde{\nu}(i)\bar{\nu}(j)$ Example: Spin-down Proton ($p\downarrow$) hopping along X:$$\tilde{h}_{x,\downarrow p}(i,j) = \frac{1}{2} \left( X_{\downarrow p, i} Y_{\downarrow p, j} + Y_{\downarrow p, i} X_{\downarrow p, j} \right) Z_{\uparrow n, i} Z_{\downarrow n, i} Z_{i'} Z_{i''} Z_{\uparrow p, j}$$

The number operator maps simply:
$$\tilde{N}_\sigma(i) = \frac{1}{2}(1 - Z_{\sigma, i})$$

so




The free pion Hamiltonian can be encoded by noting that
$$\pi_I^2(x) \rightarrow P^2 + 2PQ \sum_{m=0}^{n_b-1} 2^m Z_{I,x}^{(m)} + Q^2 \sum_{m,m'=0}^{n_b-1} 2^{m+m'} Z_{I,x}^{(m)} Z_{I,x}^{(m')}$$

And, by using the forward finite difference, we can encode
$$(\nabla \pi_I(x))^2 \rightarrow \frac{Q^2}{a_L^2} \sum_{j=1,2,3} \left[ \sum_{m,m'=0}^{n_b-1} 2^{m+m'} \left( Z_{I,x}^{(m)}Z_{I,x}^{(m')} + Z_{I,x+a_L\hat{n}_j}^{(m)}Z_{I,x+a_L\hat{n}_j}^{(m')} \right) - \sum_{m,n=0}^{n_b-1} 2^{m+n+1} Z_{I,x}^{(m)} Z_{I,x+a_L\hat{n}_j}^{(n)} \right]$$

The Axial-Vector Term can be encoded by 
$$\frac{\pi_1(x+a_L \hat{n}_1) - \pi_1(x)}{a_L} \left( a_{p\uparrow}^\dagger a_{n\downarrow} + a_{n\downarrow}^\dagger a_{p\uparrow} \right) \rightarrow \frac{Q}{2a_L} \left( \sum_{n=0}^{n_b-1} 2^n Z_{1, x+a_L\hat{n}_j}^{(n)} - \sum_{m=0}^{n_b-1} 2^m Z_{1,x}^{(m)} \right) \mathcal{O}_{N}$$
Where 
$$\mathcal{O}_N = X_{i}^{p\uparrow} X_{i}^{n\downarrow} + Y_{i}^{p\uparrow} Y_{i}^{n\downarrow} Z_{i}^{p\downarrow} Z_{i}^{n\uparrow}$$

The Weinberg-Tomozawa term is encoded by 
$$\pi_2(x) \Pi_3(x) \left( a_{p\uparrow}^\dagger a_{n\uparrow} + a_{n\uparrow}^\dagger a_{p\uparrow} \right) \rightarrow \left( P_1 + Q \sum_{m=0}^{n_b-1} 2^m Z_{2,x}^{(m)} \right) \left( P'_1 + Q' \sum_{l=0}^{n_b-1} 2^l Z_{3,x}^{(l)} \right) \mathcal{O}'_{N}$$
Where now 
$$\mathcal{O}'_N = X_{i}^{p\uparrow} X_{i}^{n\uparrow} + Y_{i}^{p\uparrow} Y_{i}^{n\uparrow} Z_{i}^{p\downarrow}$$

```python
class MyCustomHamiltonian(ProblemInstance):
    def __init__(self, pauli_dict):
        # We don't call super().__init__() if the base class is purely abstract
        # but we must implement its required methods.
        self.pauli_dict = pauli_dict
        self._n_qubits = self._calculate_qubits()
        
    # Calculate n_qubits by finding the highest index in the Pauli strings
    def _calculate_qubits(self):
        max_idx = 0
        for term in self.pauli_dict.keys():
            # Use regex to find all numbers in the string (e.g., 'Z0 X15' -> [0, 15])
            indices = [int(n) for n in re.findall(r'\d+', term)]
            if indices:
                max_idx = max(max_idx, max(indices))
        # Number of qubits is max index + 1
        return max_idx + 1

    # 1. Implementation of required n_qubits 
    def n_qubits(self):
        return self._n_qubits

    # 2. Implementation of required __str__
    def __str__(self):
        return "CustomPionEFT"

    # 3. The data source for PauliSparse encoding
    def get_pauli_extension(self):
        # PauliLCU usually expects the dictionary directly
        return self.pauli_dict
    def get_alpha(self):
        """Calculates the L1 norm of the coefficients (sum of absolute values)."""
        return sum(abs(coeff) for coeff in self.pauli_dict.values())
    def yield_PauliLCU_Info(self, **kwargs):
        """
        Yields (Dense_Pauli_String, Coefficient) pairs.
        Example: ('ZIIX', 0.5) for a 4-qubit system.
        """
        for p_str, coeff in self.pauli_dict.items():
            # Start with a list of 'I' characters for all qubits
            dense_list = ['I'] * self._n_qubits
            
            if p_str != "I":
                # Parse sparse string "Z0 X3"
                for factor in p_str.split():
                    gate = factor[0]        # 'Z'
                    idx  = int(factor[1:])  # 0
                    dense_list[idx] = gate
            
            # Join into a dense string like "ZIIX"
            dense_string = "".join(dense_list)
            yield (dense_string, coeff)
```

```python
## Set up toy model parameters
##

num_sites = 4

# Lattice size $a_L = 2.2$ fm. Units of coefficients are MeV.
# h = 10.58
# C = -98.23 
# D = 127.84

# Lattice size $a_L = 2.2$ fm. Units of coefficients are MeV.
h = 4.29
C = -40.19
D = 42.51

## Get 1-D Hamiltonian from PionlessEFT.py
qubit_ham = PionlessEFT.PionlessEFT_Hamiltonian_1D_jw(h, C, D, num_sites)

## Now use pyLIQTR to analyze the resources for simulating this Hamiltonian using qubitization and phase estimation

# --- EXECUTION ---

# 1. Convert your Hamiltonian to the dict format
pauli_dict = {}
for term, coeff in qubit_ham.terms.items():
    p_string = " ".join([f"{op}{idx}" for idx, op in term]) if term else "I"
    pauli_dict[p_string] = float(coeff.real)

# 2. Instantiate
custom_instance = MyCustomHamiltonian(pauli_dict)

# 3. Block Encoding (LCU)
encoding_type  = VALID_ENCODINGS.PauliLCU
block_encoding = getEncoding(encoding_type)(custom_instance)

# 4. Walk Operator for GSEE
walk_operator = QubitizedWalkOperator(block_encoding)

# 5. Resource Analysis
print(f"Total Qubits: {custom_instance.n_qubits}")
print(f"Alpha (Normalization): {block_encoding.alpha:.4f}")

print("\n--- Walk Operator Resource Analysis ---")
results = estimate_resources(walk_operator)
for key, value in results.items():
    print(f"{key}: {value}")

# 5. Perform Resource Analysis
print("--- Resource Estimate: Block Encoding ---")
resource_estimate_block = estimate_resources(block_encoding)
print(resource_estimate_block)

print("\n--- Resource Estimate: Walk Operator ---")
resource_estimate_walk = estimate_resources(walk_operator)
print(resource_estimate_walk)

# 6. Extract Alpha (Normalization factor)
# Vital for translating Phase back to Energy: E = alpha * cos(theta)
print(f"\nNormalization Factor (Alpha): {block_encoding.alpha}")
```
